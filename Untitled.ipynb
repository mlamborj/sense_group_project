# Predicting fishing intensity from eo variables

```python
import cartopy
import geopandas as gpd
from matplotlib import pyplot as plt
import cartopy.crs as ccrs
```

```python
gdf=gpd.read_file('./SSKN_FSRMapData_2021-2021_v1.0.0/SSKN.gdb', layer='SSKN_rfi_2010')
```

```python
# Create a Cartopy GeoAxes
ax = plt.axes(projection=ccrs.AlbersEqualArea())
# Plot the data from the shapefile
gdf.plot(ax=ax, facecolor='none', edgecolor='blue')
```
